## Groupby  and Aggregation by average

```python
tel1=PdM_telemetry1.groupby(['datetime','machineID'])[["datetime","volt","rotate","pressure","vibration"]].mean()
```

```python
print(tel1)
```

```python
new_df1 = pd.merge(tel1,error1, how='left', left_on=['datetime','machineID'], right_on = ['datetime','machineID'])
```
